```python
IP_ADDR = os.getenv("IP_ADDRESSES")
IP_ADDR = IP_ADDR.replace("\n", "").split(",")
IP_ADDR = [ip for ip in IP_ADDR]
print(IP_ADDR)
```

```python
proxies_dict_list = []


for line in IP_ADDR:
    # split the TXT format IP:PORT:USER:PASS
    ip, port, user, pwd = line.split(":")

    # create a dict directly for requests
    proxy_dict = {
        "http": f"http://{user}:{pwd}@{ip}:{port}",
        "https": f"http://{user}:{pwd}@{ip}:{port}"
    }

    proxies_dict_list.append(proxy_dict)
```

```python
address = "bc1qdm7xde6lcauqx2zy5mf2fe934pfk0yyllgehlg"
url = f"https://blockchain.info/rawaddr/{address}"

response = requests.get(url, proxies = proxies_dict_list[0], timeout=(10, 90))
response.raise_for_status()
```
